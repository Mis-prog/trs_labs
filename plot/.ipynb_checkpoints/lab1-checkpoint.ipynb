```python
import matplotlib.pyplot as plt 
import pandas as pd
from scipy.integrate import solve_ivp
import numpy as np
```

```python
n=[10,100,1000,10000]
name_z1=['z1_x','z1_v']
name_z2=['z2_x','z2_v']
name_z3=['z3_x','z3_v']

t_span = (0, 10)  # Интервал интегрирования по времени
y0 = [0, 0.09]      # Начальные условия: x(0) = 0, v(0) = 0.09
def f(t, y):
    x, v = y
    dxdt = v
    dvdt = -(8/9.1) * x * np.exp(x * x)
    return [dxdt, dvdt]
```

# Задание 1


```python
name_zad_1_x=['../output/lab_1/'+name_z1[0]+str(i)+'.txt' for i in n]
name_zad_1_v=['../output/lab_1/'+name_z1[1]+str(i)+'.txt' for i in n]
data_1_x=[pd.read_csv(name,sep=" ") for name in name_zad_1_x]
data_1_v=[pd.read_csv(name,sep=" ") for name in name_zad_1_v]
```

```python
fig,axs = plt.subplots(4,2,figsize=(10,10));

for i, value in enumerate(data_1_x): 
    axs[i][0].plot(value.t,value.x,label=len(value.t))
    axs[i][0].grid(True)
    axs[i][0].set_xlabel('t')
    axs[i][0].set_ylabel('x')
    axs[i][0].legend()
    
for i, value in enumerate(data_1_v): 
    axs[i][1].plot(value.t,value.v,label=len(value.t))
    axs[i][1].grid(True)
    axs[i][1].set_xlabel('t')
    axs[i][1].set_ylabel('v')
    axs[i][1].legend()
   
fig.tight_layout()
    
```

```python
fig,axs = plt.subplots(4,2,figsize=(10,10))

for i, value in enumerate(data_1_x): 
    sol = solve_ivp(f, t_span, y0, t_eval=np.linspace(t_span[0], t_span[1], len(value.t)))
    axs[i][0].plot(value.t,value.x,label=len(value.t))
    axs[i][0].plot(sol.t, sol.y[0], label='x(t)')
    axs[i][0].grid(True)
    axs[i][0].set_xlabel('t')
    axs[i][0].set_ylabel('x')
    axs[i][0].legend()
    
for i, value in enumerate(data_1_v): 
    sol = solve_ivp(f, t_span, y0, t_eval=np.linspace(t_span[0], t_span[1], len(value.t)))
    axs[i][1].plot(value.t,value.v,label=len(value.t))
    axs[i][1].plot(sol.t, sol.y[1], label='v(t)')
    axs[i][1].grid(True)
    axs[i][1].set_xlabel('t')
    axs[i][1].set_ylabel('v')
    axs[i][1].legend()
   
fig.tight_layout()
```

```python
fig,axs = plt.subplots(4,2,figsize=(10,10))

for i, value in enumerate(data_1_x): 
    sol = solve_ivp(f, t_span, y0, t_eval=np.linspace(t_span[0], t_span[1], len(value.t)))
    axs[i][0].plot(value.t,value.x-sol.y[0],label=(str(len(value.t))+"  epsilon="+str(max(abs(value.x-sol.y[0]).round(5)))))
    axs[i][0].grid(True)
    axs[i][0].set_xlabel('t')
    axs[i][0].set_ylabel('x')
    axs[i][0].legend()
    
for i, value in enumerate(data_1_v): 
    sol = solve_ivp(f, t_span, y0, t_eval=np.linspace(t_span[0], t_span[1], len(value.t)))
    axs[i][1].plot(value.t,value.v-sol.y[1],label= (str(len(value.t))+"  epsilon="+str(max(abs(value.v-sol.y[1]).round(5)))))
    axs[i][1].grid(True)
    axs[i][1].set_xlabel('t')
    axs[i][1].set_ylabel('v')
    axs[i][1].legend()
   
fig.tight_layout()
```
