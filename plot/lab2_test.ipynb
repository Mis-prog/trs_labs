```python
def exact_solution(x,t):
    return np.exp(t)*np.sinh(x)
```

```python
exact_solution(1,1)
```

```python
x_values = np.linspace(0, 1, 100)
t_values = np.linspace(0, 1, 100)

X, T = np.meshgrid(x_values, t_values)

exact_values = exact_solution(X, T)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(T,X, exact_values, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('Exact Solution')

plt.title('Exact Solution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Чтение данных из файла
data = np.genfromtxt("../output/lab_2/EFDS-100.txt")

# Создание сетки значений времени и координат
t_values = np.linspace(0, 1, data.shape[1])
x_values = np.linspace(0, 1, data.shape[0])

# Создание сетки значений времени и координат для построения 3D графика
T, X = np.meshgrid(t_values, x_values)

# Построение 3D графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,T, data, cmap='viridis')

# Настройка меток на осях
ax.set_xlabel('Время')
ax.set_ylabel('Координаты')
ax.set_zlabel('Значения')

# Отображение графика
plt.show()
```
