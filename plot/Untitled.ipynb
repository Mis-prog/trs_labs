```python
import numpy as np 
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

# Task 1

```python
data=np.loadtxt('../output/lab_3/task1.txt')
```

```python
size = data.shape

y_dim = size[1]
x_dim = size[0]

u1_values = data[0:].reshape(x_dim, y_dim)
x = np.linspace(0, 1, x_dim)
y = np.linspace(0, 1, y_dim)
X, Y = np.meshgrid(x, y)

fig = plt.figure(figsize = (14,9))
ax = plt.axes(projection = '3d')
ax.plot_wireframe(X,Y,u1_values, cmap = 'viridis')
```
