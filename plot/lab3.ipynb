```python
import numpy as np 
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
path="../output/lab_3/"
value=[4, 10, 20, 30, 40, 50, 60]
```

# Exact solution

```python
def u(x,y):
    return np.sin(np.pi*x)*np.sin(np.pi*x)*np.sin(np.pi*y)
```

```python
x_values = np.linspace(0, 1, 1000)
y_values = np.linspace(0, 1, 1000)

X, Y = np.meshgrid(x_values, y_values)

exact_values = u(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, exact_values, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Exact Solution')

plt.title('Exact Solution')
plt.show()
```

# Task 1

```python
fig, axs = plt.subplots(7,2, figsize=(17, 27))



for i, index in enumerate(value,1):
    data = np.genfromtxt(path+"task1_" + str(index) + ".txt")

    t_values = np.linspace(0, 1, data.shape[1])
    x_values = np.linspace(0, 1, data.shape[0])

    T, X = np.meshgrid(t_values, x_values)

    ax = fig.add_subplot(7, 2, i*2-1, projection='3d') 
    ax.set_title(f'Numerical solution for N = {index}')

    ax.plot_surface(T, X, data,cmap='viridis')


for i, index in enumerate(value,1):
    data = np.genfromtxt(path+"task1_error_" + str(index) + ".txt")

    t_values = np.linspace(0, 1, data.shape[1])
    x_values = np.linspace(0, 1, data.shape[0])

    T, X = np.meshgrid(t_values, x_values)
    ax = fig.add_subplot(7, 2, i*2 , projection='3d')
    ax.set_title(f'Error for N = {index}')
    ax.plot_surface(T, X, data, cmap='viridis',label=str(np.max(data)))
    ax.legend()

plt.show()


```
