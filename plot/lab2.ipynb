```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

# Задание 1

## Явная схема

```python
def exact_solution(x,t):
    return np.exp(t)*np.sinh(x)
```

```python
x_values = np.linspace(0, 1, 100)
t_values = np.linspace(0, 0.5, 100)

X, T = np.meshgrid(x_values, t_values)

exact_values = exact_solution(X, T)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, T, exact_values, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('Exact Solution')

plt.title('Exact Solution')
plt.show()
```
